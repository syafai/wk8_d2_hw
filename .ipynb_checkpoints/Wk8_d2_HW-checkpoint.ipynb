Design a flexible OOP system to describe characters in a RPG game

There are 4 Unique Characters which use Different Types of Attacking mechanisms and Defending mechanisms.

Orges - fights with club and defends with shield

Peons - fights with club and defends with shield

Knights - fights with a Sword and defends with Armor

Archer - fights with Bow and Arrow and has only is tunic to protect them

All Characters can collect gold and this will always be the same for every new type of character

At the end of this each character should be able to attack and defend and collect gold. If needed any character should be able to change his fighting style, say if a knight losing his sword, but finds a club he should be able to change his fighting mechanism to use the club

Hint: To complete this assignment you will need multiple classes and interfaces and an Abstract class

```typescript
// Abstract class representing a character
abstract class Character {
  public gold: number = 0;

  abstract attack(): void;
  abstract defend(): void;

  collectGold(amount: number): void {
    this.gold += amount;
  }
}

// Interface for characters with attacking mechanism
interface Attacker {
  attack(): void;
}

// Interface for characters with defending mechanism
interface Defender {
  defend(): void;
}

// Class representing the Ogres character
class Ogre extends Character implements Attacker, Defender {
  private weapon: string = "club";
  private shield: string = "shield";

  attack(): void {
    console.log(`Ogre attacks with ${this.weapon}`);
  }

  defend(): void {
    console.log(`Ogre defends with ${this.shield}`);
  }
}

// Class representing the Peons character
class Peon extends Character implements Attacker, Defender {
  private weapon: string = "club";
  private shield: string = "shield";

  attack(): void {
    console.log(`Peon attacks with ${this.weapon}`);
  }

  defend(): void {
    console.log(`Peon defends with ${this.shield}`);
  }
}

// Class representing the Knights character
class Knight extends Character implements Attacker, Defender {
  private weapon: string = "sword";
  private armor: string = "armor";

  attack(): void {
    console.log(`Knight attacks with ${this.weapon}`);
  }

  defend(): void {
    console.log(`Knight defends with ${this.armor}`);
  }
}

// Class representing the Archer character
class Archer extends Character implements Attacker {
  private weapon: string = "bow and arrow";
  private tunic: string = "tunic";

  attack(): void {
    console.log(`Archer attacks with ${this.weapon}`);
  }

  defend(): void {
    console.log(`Archer defends with ${this.tunic}`);
  }
}


```

```typescript
// Usage example:
let ogre = new Ogre();
ogre.attack(); // Output: Ogre attacks with club
ogre.defend(); // Output: Ogre defends with shield
ogre.collectGold(10);
console.log(`Ogre's gold: ${ogre.gold}`); // Output: Ogre's gold: 10

let knight = new Knight();
knight.attack(); // Output: Knight attacks with sword
knight.defend(); // Output: Knight defends with armor
knight.collectGold(5);
console.log(`Knight's gold: ${knight.gold}`); // Output: Knight's gold: 5

// Changing Knight's fighting mechanism
knight.attack = function () {
  console.log(`Knight attacks with club`);
};
knight.attack(); // Output: Knight attacks with club
```
